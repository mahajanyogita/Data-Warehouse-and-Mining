```python
l=list(map(int,input("Enter List of Numbers: ").split()))
k=int(input("Enter K : "))
klist=[]
for i in range(k):
    klist.append(l[i])
    
while(True):
    clusters=[]
    for d in klist:
        clusters.append([int(d)])
    
    for d in l:
        newlist=[]
        for j in klist:
            newlist.append(abs(j-d))
        
        clusters[newlist.index(min(newlist))].append(int(d))
        print(clusters)
    
    newklist=[]
    for i in range(k):
        newklist.append(sum(list(clusters[i]))/len(clusters[i]))
        
    if(newklist==klist):
        break
    else:
        klist=newklist
        
for i in range(k):
    if klist[i] not in l:
        clusters[i].remove(int(klist[i]))
        
        
```
